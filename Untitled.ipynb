```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('u.data',sep='\t',names = ['user_id','item_id','rating','ts'])
```

```python
df.head()
```

```python
cols = "item_id|title| release date | video release date |\
              IMDb URL | unknown | Action | Adventure | Animation |\
              Children's | Comedy | Crime | Documentary | Drama | Fantasy |\
              Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi |\
              Thriller | War | Western ".split('|')
movies = pd.read_csv('u.item',sep='|',names = cols,encoding='ISO-8859-1')
movies.head()
```

```python
 movies[['item_id','title']].head()
```

```python
data = pd.merge(df.drop('ts',axis=1), movies[['item_id','title']], on='item_id')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['title'].nunique()
```

```python
data['user_id'].nunique()
```

```python
movies=data['title'].unique()
for n in movies:
    print(n,' '*5,data['rating'][data['title']==n].mean())
```

another way to find the rating is 

```python
avg_rates = data.groupby('title')['rating'].mean()
avg_rates.head(20)
```

```python
avg_rates.sort_values(ascending=False).head(20)
```

```python
rates_count = data.groupby('title')['rating'].count()
rates_count.head(20)
```

```python
rates_count.sort_values(ascending=False).head(20)
```

```python
plt.figure(figsize =(18,5))
rates_count.hist(bins=50)
t = plt.xticks(range(0,601,20))
```

```python
plt.figure(figsize=(18,5))
avg_rates.hist(bins=50)
```

```python
plt.scatter(x=avg_rates,y=rates_count)
plt.xlabel('Avg Rating')
plt.ylabel('no. of people rated')
plt.grid()
```

```python
df_pivot=data.pivot_table(index='user_id',columns='title',values='rating')
df_pivot.head()
```

```python
inp = 'Star Wars (1977)'
```

```python
df_pivot[inp].head()
```

```python
sim_inp = df_pivot.corrwith(df_pivot[inp])
```

```python
sim_inp
```

```python
sin_dif = pd.DataFrame(sim_inp,columns=['Correlation'])
sin_dif.head()
```

```python
sin_dif.sort_values(by = 'Correlation',ascending=False).head()
```

```python
sin_dif['count'] = rates_count
```

```python
sin_dif['avg_rate'] = avg_rates
```

```python
sin_dif.head()
```

```python
sin_dif[sin_dif['count']>100].sort_values('Correlation',ascending=False).head(10)
```

```python
recom = sin_dif[sin_dif['count']>100].sort_values('Correlation',ascending=False)
recom.drop(inp,axis=0,inplace=True)# it remove the star wars
recommended = recom.index[:3]
for r in recommended:
    print(r) #this function is used  to select if the person want to see after any star wars movie 
```
